## Ans1

The three measures of central tendency are: Mean , Median and Mode

## Ans2

Mean is used to calculate the average of the data<br> Mean = (Sum of all elements/ Total number of elements)<br> 
Median is used to find the middle most datapoint of the data<br>(Element at the center of sorted dataset)<br>
Mode is used to give the most frequently occured element from the data

## Ans3

```python
data = [178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]
import numpy as np
import statistics
```

```python
mean = np.mean(data)
median = np.median(data)
mode = statistics.mode(data)
print(f'Mean ={mean}, Median = {median}, Mode = {mode}')
```

## Ans5

Range: Range is calculated by subtracting the minimum value from the maximum value in a dataset<br>
For example : For a dataset:  50, 65, 70, 80, and 90, the range would be 40 (90 - 50).


Variance: Variance is calculated by averaging the squared differences of each value from the mean of the dataset. <br>
For example : For a dataset: 50, 65, 70, 80, and 90, the variance would be approximately 238.4. 


Standard deviation: The standard deviation is another measure of dispersion that is calculated by taking the square root of the variance.<br>
For example : For a dataset: 50, 65, 70, 80, and 90, the standard deviation would be 15.4.

## Ans14

Outliers effect Mean particularly as measure of central tendency as it requires the sum of all the values.<br>
Median and Mode arenot effected to such extent by the presence of outliers.


```python
# Example:
# Data with no ouliers:
data = [10,14,12,8,10,7,6,11]
np.mean(data)
```

```python
np.median(data)
```

```python
statistics.mode(data)
```

```python
# Data with outliers
data1 = [10,14,12,8,10,7,6,11,100]
np.mean(data1)
```

```python
np.median(data1)
```

```python
statistics.mode(data1)
```

We can see that mean of data with outliers varies significantly with mean of data with no outliers whereas median and mode remain unchanged
