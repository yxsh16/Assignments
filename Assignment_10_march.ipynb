## Ans1

Estimation statistics is a branch of statistics that deals with making inferences or predictions about population parameters (such as the mean or standard deviation) based on a sample of data. In estimation statistics, we use sample statistics to estimate population parameters.


Point estimate: A point estimate is a single value that is used to estimate an unknown population parameter. It is based on a sample statistic and is used to make an inference about the population parameter. <br>
Interval estimate: An interval estimate is a range of values that is used to estimate an unknown population parameter. It is based on a sample statistic and takes into account the level of confidence or reliability we want to have in our estimate. 

## Ans2

```python
import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    standard_error = sample_std_dev / math.sqrt(sample_size)
    margin_of_error = 1.96 * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return (sample_mean, lower_bound, upper_bound)
```

## Ans5

```python
import scipy.stats as stats

pop1 = [1, 2, 3, 4, 5]
pop2 = [3, 4, 5, 6, 7]

sample_mean1 = sum(pop1) / len(pop1)
sample_mean2 = sum(pop2) / len(pop2)

sample_std1 = stats.tstd(pop1)
sample_std2 = stats.tstd(pop2)

pooled_std = ((len(pop1) - 1) * sample_std1 ** 2 + (len(pop2) - 1) * sample_std2 ** 2) / (len(pop1) + len(pop2) - 2)
pooled_std = pooled_std ** 0.5

t_stat = (sample_mean1 - sample_mean2) / (pooled_std * (2 / len(pop1)))
p_value = stats.t.sf(abs(t_stat), len(pop1) + len(pop2) - 2) * 2

alpha = 0.05


if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

## Ans14

A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

```python
import numpy as np
from scipy.stats import t


null_mean_diff = 10

sample_mean = 8
sample_std = 3
sample_size = 100


t_statistic = (sample_mean - null_mean_diff) / (sample_std / np.sqrt(sample_size))
degrees_freedom = sample_size - 1
p_value = t.sf(np.abs(t_statistic), degrees_freedom) * 2

alpha = 0.05

if p_value < alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
    
print("t-statistic: {:.2f}".format(t_statistic))
print("p-value: {:.4f}".format(p_value))
```

## Ans16

Two groups of students are given different study materials to prepare for a test. The first group (n1 =
30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
groups are equal with a significance level of 0.01.

```python
n1 = 30
mean1 = 80
std1 = 10

n2 = 40
mean2 = 75
std2 = 8


s_pooled = np.sqrt(((n1-1)*(std1**2) + (n2-1)*(std2**2)) / (n1+n2-2))
t_statistic = (mean1 - mean2) / (s_pooled * np.sqrt(1/n1 + 1/n2))
degrees_freedom = n1 + n2 - 2
p_value = 2 * (1 - t.cdf(abs(t_statistic), degrees_freedom))


alpha = 0.01

if p_value < alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
    
print("t-statistic: {:.2f}".format(t_statistic))
print("p-value: {:.4f}".format(p_value))
```

```python
import numpy as np
from scipy.stats import t

n = 50
mean = 4
std = 1.5

confidence_level = 0.99

degrees_of_freedom = n - 1
t_value = t.ppf(1 - (1 - confidence_level) / 2, degrees_of_freedom)

margin_of_error = t_value * std / np.sqrt(n)

lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print("99% confidence interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```
